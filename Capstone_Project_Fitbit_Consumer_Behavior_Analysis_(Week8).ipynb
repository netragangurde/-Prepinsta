### Importing the libraries and loading the dataset

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Daily activity Dataset

```python
dailyActivity_merged=pd.read_csv('/content/dailyActivity_merged.csv')
```

```python
dailyActivity_merged
```

```python
plt.figure(figsize=(15, 6))
plt.plot(dailyActivity_merged['TotalSteps'], label='Daily Steps', marker='o')
plt.plot(dailyActivity_merged['Calories'], label='Daily Calories Burnt', marker='o')
plt.title('Daily Steps and Calories Burnt')
plt.xlabel('Day')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

Merging all hourly datasets

```python
df1=pd.read_csv('/hourly_merged.csv')
```

```python
df1
```

```python
from matplotlib import pyplot as plt
df1['TotalIntensity'].plot(kind='line', figsize=(8, 4), title='Total Hourly Intensity')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
df2=pd.read_csv('/content/hourlySteps_merged.csv')
```

```python
cal_step_merged=pd.merge(df1,df2,on =['Id','ActivityHour'])
```

```python
cal_step_merged
```

```python
hourly_merged=pd.merge(cal_step_merged,df3,on =['Id','ActivityHour'])
```
